```python
import cv2
import matplotlib.pyplot as plt

# Load and resize the image to reduce memory usage
image = cv2.imread('C:\\Users\\Itish\\Downloads\\PRJ Face Detection\\Data Set\\people 16.jpg')
image = cv2.resize(image, (800, 600))  # Resize for better memory management

try:
    # Run the detector on the resized image
    detections = cnn_detector(image, 1)
    
    for face in detections:
        l, t, r, b, c = face.rect.left(), face.rect.top(), face.rect.right(), face.rect.bottom(), face.confidence
        print(c)
        cv2.rectangle(image, (l, t), (r, b), (255, 255, 0), 2)

    # Display the image with detections
    cv2.imshow('Detections', image)
    cv2.waitKey(10000)

except cv2.error as e:
    print("OpenCV Error:", e)
except MemoryError as e:
    print("Memory Error:", e)
finally:
    cv2.destroyAllWindows()  # Clean up OpenCV windows
    del image  # Optionally free memory
```
